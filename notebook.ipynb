```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import mode 

# Read in the Netflix CSV as a DataFrame
netflix_series = pd.read_csv("netflix_data.csv")
netflix_df = pd.DataFrame(netflix_series)
netflix_df.head(20)
```

```python
net_90s = (netflix_df["release_year"] >= 1990) & (netflix_df["release_year"] <= 1999)

movies_90s = netflix_df[net_90s]

print(movies_90s)

np_netflix = np.array(movies_90s["duration"])
duration = mode(np_netflix)


np_netflix = np.array(movies_90s["duration"])
duration_mode = mode(np_netflix)
duration = int(duration_mode.mode[0])
plt.hist(np_netflix, bins=10)
plt.xlabel("Duração (minutos)")
plt.ylabel("Frequência")
plt.title("Distribuição das Durações dos Filmes da Década de 90")
plt.show()

print(f"A duração mais comum dos filmes da década de 90 é {duration_mode.mode[0]} minutos.")
print(f"Ela aparece {duration_mode.count[0]} vezes nos dados.")

```

```python

net_90s_action = (netflix_df["release_year"] >= 1990) & (netflix_df["release_year"] <= 1999) & (netflix_df["genre"] == "Action")
movies_90s_action = netflix_df[net_90s_action]


np_netflix_action = np.array(movies_90s_action["duration"])


short_movie_count = np.sum(np_netflix_action < 90)
print(f"O número de filmes de ação com duração menor do que 90 minutos lançados desde a década de 90 é {short_movie_count}.")
```
